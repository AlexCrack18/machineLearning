```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
import statsmodels.api as sm
from statsmodels.stats.outliers_influence \
import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm
from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize , poly)
```

```python
Auto = load_data('Auto')
Auto.columns
```

```python
# Create intercept by adding a column of ones
# At the same time use horspower as the predictor
X = pd.DataFrame({'intercept': np.ones(Auto.shape[0]), 'horsepower': Auto['horsepower']})

# Create the y
y = Auto['mpg']

# use the sm.OLS(y, X) to create the model
model = sm.OLS(y, X)

# Use the fit method to fit the model
results = model.fit()

# Then summarize the model
summarize(results)
```

```python
design = MS(['horsepower'])
temp = design.fit_transform(Auto)

newDataFrame = pd.DataFrame({'horsepower': [70, 120, 160, 200]})
newX = design.transform(newDataFrame)
newY = results.get_prediction(newX)
print(newY.predicted_mean, '\n')
newY.conf_int(obs=True, alpha=0.05)
```
